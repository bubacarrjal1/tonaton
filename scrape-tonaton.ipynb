```python
import requests
from bs4 import BeautifulSoup
import time
import csv

# Make a GET request to fetch the raw HTML content
source = requests.get('https://tonaton.com/en/ads/ghana/land').text

# parse the html content
soup = BeautifulSoup(source, "lxml")
# print(soup.prettify())

csv_file = open('tonaton_scrape.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['title', 'size', 'region', 'price'])

for info in soup.find_all('div', class_="container--2uFyv"):
    title = info.find('span', class_="title--3yncE").text
    print(title)

    size = info.find('div', class_="content--3JNQz").div.div.text
    print(size)

    region = info.find('div', class_="description--2-ez3").text
    print(region)

    price = info.find('div', class_="price--3SnqI color--t0tGX").span.text
    print(price)
    
    print()
    
    csv_writer.writerow(['title', 'size', 'region', 'price'])
    
csv_file.close()
```
